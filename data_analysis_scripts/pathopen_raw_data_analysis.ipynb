#### Image Magnification Breakdown

It was reported that the Pathologist CK didn't include the magnification of the eyepience lens (10x) while reporting the resolution of the images where pathologist AC included all those. So, to make this data consistent, resolutions collected by CK are being multiplied with 10 to ensure that resolution reflects the objective magnification of the microscope multiplied by the eyepiece lens magnification

Image Magnifications Reported by AC

```python
vqa_data[vqa_data['Pathologist ID'] == 'AC'][['Image1_Mag', 'Image2_Mag', 'Image3_Mag']].apply(pd.Series.value_counts)
```

Image Magnifications Reported by CK

```python
vqa_data[vqa_data['Pathologist ID'] == 'CK'][['Image1_Mag', 'Image2_Mag', 'Image3_Mag']].apply(pd.Series.value_counts)
```

This shows that CK did not incorporate the magnification of eyepiece lens (10x) when reporting the overall image resolution. So multiplying all the magnifications reported by CK by 10 will make our data consistent

```python
def convert_mag(mag):
    path_id = mag.iloc[0]
    image_mag = mag.iloc[1]
    if pd.isnull(image_mag):
        return image_mag
    
    image_mag = image_mag.lower().strip()
    if path_id == 'CK':    
        if 'x' in image_mag:
            return str(int(image_mag.split('x')[0]) * 10) + 'x'
    return image_mag
```

```python
vqa_data['Image1_Mag'] = vqa_data[['Pathologist ID', 'Image1_Mag']].apply(convert_mag, axis=1)
vqa_data['Image2_Mag'] = vqa_data[['Pathologist ID', 'Image2_Mag']].apply(convert_mag, axis=1)
vqa_data['Image3_Mag'] = vqa_data[['Pathologist ID', 'Image3_Mag']].apply(convert_mag, axis=1)
```

```python
image1_mag_counts = vqa_data['Image1_Mag'].value_counts()
image2_mag_counts = vqa_data['Image2_Mag'].value_counts()
image3_mag_counts = vqa_data['Image3_Mag'].value_counts()

image_mag_counts = image1_mag_counts.add(image2_mag_counts, fill_value=0).add(image3_mag_counts, fill_value=0).sort_values(ascending=False)

ax = image_mag_counts.plot(kind='bar', figsize=(8, 5), title='Image Magnification Breakdown', xlabel='Image Magnification', ylabel='Number of Cases')
ax.bar_label(ax.containers[0], padding=2)
ax.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig(f'{output_data_path}/pathopen_image_magnification_breakdown.png')
plt.show()
```

```python
open_ended_questions = list(vqa_data['Open Ended - Question 1'].dropna()) + list(vqa_data['Open Ended - Question 2'].dropna())
open_ended_correct_answers = list(vqa_data['Open Ended - Answer 1'].dropna()) + list(vqa_data['Open Ended - Answer 2'].dropna())
open_ended_wrong_answers = list(vqa_data['Open Ended - Wrong Answer 1'].dropna()) + list(vqa_data['Open Ended - Wrong Answer 2'].dropna())

open_ended_questions_words_5 = [sample for sample in open_ended_questions if len(sample.split()) >= 5]
open_ended_correct_answers_words_5 = [sample for sample in open_ended_correct_answers if len(sample.split()) >= 5]
open_ended_wrong_answers_words_5 = [sample for sample in open_ended_wrong_answers if len(sample.split()) >= 5]

open_ended_questions_words_10 = [sample for sample in open_ended_questions if len(sample.split()) >= 10]
open_ended_correct_answers_words_10 = [sample for sample in open_ended_correct_answers if len(sample.split()) >= 10]
open_ended_wrong_answers_words_10 = [sample for sample in open_ended_wrong_answers if len(sample.split()) >= 10]

print(f'Number of Open-Ended Questions (Words >= 5): {len(open_ended_questions_words_5)}/{len(open_ended_questions)}')
print(f'Number of Open-Ended Correct Answers (Words >= 5): {len(open_ended_correct_answers_words_5)}/{len(open_ended_correct_answers)}')
print(f'Number of Open-Ended Wrong Answers (Words >= 5): {len(open_ended_wrong_answers_words_5)}/{len(open_ended_wrong_answers)}')

print(f'Number of Open-Ended Questions (Words >= 10): {len(open_ended_questions_words_10)}/{len(open_ended_questions)}')
print(f'Number of Open-Ended Correct Answers (Words >= 10): {len(open_ended_correct_answers_words_10)}/{len(open_ended_correct_answers)}')
print(f'Number of Open-Ended Wrong Answers (Words >= 10): {len(open_ended_wrong_answers_words_10)}/{len(open_ended_wrong_answers)}')
```
